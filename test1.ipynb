```python
def перевірка_препаратів(препарати):
    for препарат in препарати:
        if type(препарат['кількість']) != int:
            print(препарат['назва'] + ": Помилка даних")
            continue
        
        if type(препарат['температура']) != float:
            print(препарат['назва'] + ": Помилка даних")
            continue
        
        if препарат['температура'] < 5:
            температура = "Надто холодно"
        elif препарат['температура'] > 25:
            температура = "Надто жарко"
        else:
            температура = "Норма"
        
        match препарат['категорія']:
            case 'antibiotic':
                категорія = "Рецептурний препарат"
            case 'vitamin':
                категорія = "Вільний продаж"
            case 'vaccine':
                категорія = "Потребує спецзберігання"
            case _:
                категорія = "Невідома категорія"
        
        print(препарат['назва'] + ": " + категорія + ", " + температура)


препарати = [
    {"назва": "Амоксицилін", "кількість": 100, "категорія": "antibiotic", "температура": 4.0},
    {"назва": "Вітамін C", "кількість": 200, "категорія": "vitamin", "температура": 10.5},
    {"назва": "Щеплення від грипу", "кількість": 50, "категорія": "vaccine", "температура": 2.0},

    {"назва": "Біотин", "кількість": "не вказано", "категорія": "vitamin", "температура": 22.0},
    {"назва": "Невідомий препарат", "кількість": 30, "категорія": "unknown", "температура": 30.0}
]

перевірка_препаратів(препарати)
```
